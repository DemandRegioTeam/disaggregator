```python
par = pd.DataFrame.from_dict(config.gas_load_profile_parameters_dict())
districts = ['11000']
df = temp_df[[x for x in districts]]
```

```python
df.head()
```

```python
slp = 'KO'
```

```python
par_slp = par.loc[(par.index == slp)].reset_index()
A = par_slp['A'][0]
B = par_slp['B'][0]
C = par_slp['C'][0]
D = par_slp['D'][0]
mH = par_slp['mH'][0]
bH = par_slp['bH'][0]
mW = par_slp['mW'][0]
bW = par_slp['bW'][0]
```

```python
for landkreis in districts:
        te = temp_df[landkreis].values
        for i in range(len(te)):
            df[landkreis][i] = ((A / (1 + pow(B / (te[i] - 40), C)) + D)
                                + max(mH * te[i] + bH, mW * te[i] + bW))
```
